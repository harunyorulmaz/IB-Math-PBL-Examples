```python
# Premier League veri analizi
import pandas as pd
import matplotlib.pyplot as plt

# Ã–rnek veri
data = pd.DataFrame({
    "Team_Age": [25.2, 27.1, 24.3, 28.4],
    "Wins": [15, 18, 12, 20]
})

# Scatter plot
data.plot.scatter(x="Team_Age", y="Wins")
plt.title("Team Age vs Wins in Premier League")
plt.xlabel("Team Age")
plt.ylabel("Wins")
plt.show()
```
