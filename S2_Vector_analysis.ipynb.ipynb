```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


v1 = np.array([1, 2, 3])
v2 = np.array([2, 4, 6])
v3 = np.array([1, 1, 0])
v4 = np.array([0, 1, 1])


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')


ax.quiver(0, 0, 0, v1[0], v1[1], v1[2],
          color='r', arrow_length_ratio=0.1, label='v1=(1,2,3)', linewidth=2)
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2],
          color='b', arrow_length_ratio=0.1, label='v2=(2,4,6)', linestyle='dashed')
ax.quiver(0, 0, 0, v3[0], v3[1], v3[2],
          color='g', arrow_length_ratio=0.1, label='v3=(1,1,0)', linewidth=3)
ax.quiver(0, 0, 0, v4[0], v4[1], v4[2],
          color='m', arrow_length_ratio=0.1, label='v4=(0,1,1)', linewidth=3)


max_val = max(np.concatenate([v1, v2, v3, v4])) + 1
ax.set_xlim([0, max_val])
ax.set_ylim([0, max_val])
ax.set_zlim([0, max_val])
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Representation of Four Vectors', fontsize=14)

# Grid ve lejant
ax.grid(True)
plt.legend()


matrix = np.array([v1, v2, v3, v4])
rank = np.linalg.matrix_rank(matrix)
plt.figtext(0.1, 0.02, f"Matrix Rank: {rank} (If rank<4, there are linearly dependent vectors)",
            ha='left', fontsize=10, style='italic')

plt.tight_layout()
plt.show()
```
